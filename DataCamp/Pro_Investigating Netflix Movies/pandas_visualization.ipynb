```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
np.random.seed(123456)
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2020', periods=1000))
ts
```

```python
ts = ts.cumsum()
ts
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 6), index=ts.index, columns=list('ABCDEF'))
df
```
