# `__name__` values:

1. If you have imported a module, then `__name__` on that module has a string value => the name of the module file.  So if you `import mymod`, then `mymod.__name__` will be `mymod`
2. The first thing that Python runs when you start a Python program has a special value for `__name__` -- `"__main__"`.
