```python
while True: 
    user_input = input('Date: ')

    month_list = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December"
]

    if "/" in user_input: 
        input_parts = user_input.split("/")
        try: 
            month = int(input_parts[0])
            if month <1 or month > 12:  # check for mm is more than 12 month, or less than 1,
                continue
            
            day = int(input_parts[1]) 
            if day <1 or day > 31:      # check for dd is more than 31 days, or less than 1 
                continue
        
            year = int(input_parts[2])

        except ValueError: 
            continue


    elif ", " in user_input:
        input_parts = user_input.split(" ")
        month_position = input_parts[0].capitalize()
        try: 
            month = month_list.index(month_position) + 1
        except ValueError:
            continue
 
        try:
            day =  int(input_parts[1].strip(","))
            if day <1 or day > 31:      # check for dd is more than 31 days, or less than 1 
                continue
        
            year = int(input_parts[2])
        except ValueError:
            continue
    
    else:
        continue
    
    break
print (f'{year}-{month:02}-{day:02}')        
 
```
